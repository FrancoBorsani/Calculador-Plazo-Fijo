```python
def valor_futuro(vp, i):
    return vp*i*(1/12)
```

```python
valor_presente = 220000
interes_anual = 37
periodos = 12
```

```python
print(valor_futuro(valor_presente, (interes_anual/100)))
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# x axis values
x = []
y = []
i = 1
j = 1
while(i <= periodos):
    x.append(i)
    i=i+1

valor_acumulado = valor_presente

while(j <= periodos):
    valor_acumulado = valor_acumulado + valor_futuro(valor_presente, (interes_anual/100))
    y.append(valor_acumulado)
    j=j+1
    
f= plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
    
# plotting the points 
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 1,
         marker='o', markerfacecolor='blue', markersize=5)
  
    
# setting x and y axis range
plt.xlim(1, (periodos + 1))
  
# naming the x axis
plt.xlabel('x - Meses')
# naming the y axis
plt.ylabel('y - Monto')
  
# giving a title to my graph
plt.title("Calculador plazo fijo - tasa anual: " + str(interes_anual) + "%")


for a,b in zip(x, y): 
    plt.text(a, b, str(b))

  

  
# function to show the plot
plt.show()

print("Monto obtenido al final el plazo fijo de " + str(periodos) + " meses: " + str(y.pop()))
print("Intereses ganados: " + str(valor_futuro(valor_presente, (interes_anual/100) * periodos )) )
```
